The project examines how the preliminary characteristics of various counties influence voting behaviour, whether individuals vote for Republicans or Democrats, during the United States election of 2020. The total population and female population proportion are among the preliminary characteristics. The analysis includes four datasets obtained from [main](https://www.kaggle.com/datasets/unanimad/us-election-2020) and [additional](https://www.kaggle.com/pauldesalvo/2020-election-analysis/data?select=president_counties.csv) and [additional](https://data.census.gov/cedsci/table?q=B19013&g=0400000US06%240500000_04000E0US06&tid=ACSDT5Y2019.B19013&moe=false&tp=true.). Also, the project includes the API-based web scraping from the Census Bureau, American Community Survey [link](https://api.census.gov/data/2020/acs/acs5/variables.html).

The primary dataset, US Election 2020, contains 32177 observations and six variables:  
- ***state***: state of the United State.
- ***county***: county of the corresponding state.
- ***candidate***: presidential candidate.
- ***party***: party of the presidential candidate.
- ***total_votes***: number of votes that the presidential candidate got in the county in the corresponding state.
- ***won***: a boolean expression indicating whether the presidential candidate won in in the county in the corresponding state (***True*** = *won* and ***False*** = *lost*).



The additional dataset, 2020 Election analysis, contains 3045 observations and 17 variables, from which six comes from the primary dataset, *state*, *county*, *candidate*, *party*, *total_votes*, and *won*:
- ***fips***: numbers that are used to designate certain geographic locations.
- ***state_code***: code of the state.
- ***male***: number of males in the county in the corresponding state.
- ***female***: number of females in the county in the corresponding state. 
- ***median_age***: median age in the county in the corresponding state.
- ***population***: population in the county in the corresponding state. 
- ***female_percentage***: percantage of female population in the county in the corresponding state.
- ***lat***: latitude coordinates of the county.
- ***long***: longitude coordinates of the county. 
- ***color***: colour of the party (Republicans = red and Democrats = blue).

The additional dataset contains a smaller amount of observations since it includes only the won counties of each party. However, it is used to extract preliminary characteristics of the county. 


The additional dataset, American Community Survey of households (2019), contains 58 observations and 4 variables:
- ***median_income***: median income.
- ***marginal_error***: marginal error.
- ***id***: some legislation number.
- ***geographic_area_name***: counties of California.

The data analysis suggests that the preliminary counties' characteristics, such as female population proportion and total population (independent variables), have explanatory power on voting behaviour (dependent variable). The analysis includes fundamental economic variables: the median income for each state and county and the unemployment rate for each county. The preliminary research suggests that counties' characteristics, median age, population, and per capita income have explanatory power over voting behaviour.  

In Project One, we identified two independent variables that might explain the variation in Republican vote share; county female percentage, median age, and population; independent variables differed between Republican and Democrats counties. The main finding was that females are more likely to vote for Democrats than Republicans, while males are more likely to vote for Republicans than Democrats. The median age was higher for Republican Party. However, these variables might explain an insignificant portion of a variation in the dependent variable; hence, including additional variables might benefit our analysis, for example, income per capita. We started descriptive statistics analysis from one state, California. We have found an associative relationship between income and Republicans' vote share in California; as per capita income increases, the Republicans' vote share decreases. However, we stated that it is descriptive statistics, and we can not make any causal inference from it. Also, we split the dataset into two subsamples, Low-Income and High-Income. We found that the Republicans' vote share is different, suggesting that the per capita income is a good candidate for the model. Similarly, few demographic characteristics differed. On the other hand, we found that the median age and female percentage are not explanatory variables since both variables did not significantly change when we split the dataset. 

We moved to the regression analysis and started choosing the new best variables for our model. The correlation table explains the linear relationship between independent and dependent variables. Furthermore, we incorporated a stepwise forward selection function based on different measures; Adjusted R-square, Akaike Information Criteria (AIC), and Mean Squared Error (MSE). The stepwise selection suggested including all variables, excluding some demographic characteristics. Overall, the model showed no relationship between economic variables, per capita income and the unemployment rate, with the dependent variable, Republicans' vote share. However, as always with observable data, it is hard to identify the causal inference; hence, we might expect that the model might be biased. 

In the final section, we compared performance with simple linear regression using a machine learning approach called regression tree. The MSE for the regression tree is lower than for simple linear regression.  

Ultimately, we found no relationship between the per capita income and Republicans' vote share. ***However, it is crucial not to exaggerate the precision of statements because it is challenging to find causality with observable data.*** The potential biases still might be presented; omitted variable bias, open backdoor paths, and reverse correlation.

```python
# The subset of the dataset: accessing rows with REP or DEM party.
df_president_short = df_president[df_president\
                                  .party.isin(["REP","DEM"]) == True]

# Extracting variables for data cleaning: state, county, party, and total_votes.
df_president_short = df_president_short[["state","county",\
                                         "party","total_votes"]]

# Cleaning the data; switching rows to columns. 
df_president_short["REP_votes"] = df_president_short[df_president_short.\
                                                     party == "REP"]["total_votes"]
df_president_short["DEM_votes"] = df_president_short[df_president_short.\
                                                     party == "DEM"]["total_votes"]

# Grouping the data. Mainly, we want to remove NAs in rows since REP total_votes does not have DEM total_votes.
df_president_short = df_president_short.groupby(["county"],
                                                as_index = False).sum()

# Creating new columns; percent of votes of each party from total votes.
df_president_short["REP_share"] = df_president_short["REP_votes"] / df_president_short["total_votes"] * 100
df_president_short["DEM_share"] = df_president_short["DEM_votes"] / df_president_short["total_votes"] * 100

# Rounding percent of votes.
df_president_short["REP_share"] = df_president_short["REP_share"]\
.round(decimals = 5)

df_president_short["DEM_share"] = df_president_short["DEM_share"]\
.round(decimals = 5)

# Changing the type of variables type from float to int
df_president_short = df_president_short.astype({"DEM_votes":'int',
                                                "REP_votes":'int'})
```

```python
fig, ax1 = plt.subplots(figsize = (12,6))

ax1.hist(df_president_short["REP_share"], bins = 80, color= ['#FF000099'])
plt.setp(ax1, ylabel='Frequency')
plt.xlim([0,100])

fig.suptitle("Histogram: Total Votes Share for Republicans (county-level)",\
             fontsize = 20, fontweight = 'bold')

plt.show()
```

The histogram above presents the total votes share for Republicans (by county); it is worth noting that this figure would be mirrored for Democrats, as votes that did not go to Republicans went to Democrats; thus, we did not plot the Democratic Party. According to the histogram, the Republican Party receives 60% of the vote on average. Democrats receive 40%, on average. The distribution might be bell-shaped, normally distributed. However, even though it indicates that Republicans received 60% of total votes on average, it does not mean that they received the majority of votes overall because the county's population plays a significant impact. In other words, the Republican Party might get 60% of the total votes in a county with 10,000 people, while Democrats received 70% of the total votes in a county with 1,000,000 people.

The figure above represents the total number of won counties by major parties, Democrats and Republicans, in each state. <h4> API-based Scraping </h4>

```python
# Load the API key file.
api = pd.read_csv("/Users/maximolii/Desktop/UofT/Years/Fourth Year/2/ECO225/Project/ECO225Project/Data/API.csv")

# Extract the US census API key.
census_api = api.loc[api["API"] == 'census']["KEY"].iloc[0]
```

```python
# Define function that returns the response as a DataFrame.
def json_to_dataframe(response):
    """Return the response as DataFrame.
    """
    return pd.DataFrame(response.json()[1:], columns=response.json()[0])
```

```python
# URL to the census.gov API.
# CONSTANT
HOST = "https://api.census.gov/data"

# Census API.
# CONSTANT
KEY = "&key={0}".format(census_api)

# URL struture for data extraction. 
# Year.
year = "/2020"

# Path to the required dataset.
dataset = "/acs/acs5"
g = "?get="

# Get variables required.
variables = "NAME,B19301_001E"

# Geography code.
# "state:*" -> all states.
# "us:*" -> whole US.
# In general: "state:{state code}"
location = "&for=state:*"
```

The chunk of code above represents the parts of the requested link to the Census Bureau, api.census.gov/data. The *HOST* represents the link to the API server of the census, the *KEY* represents the API key, *year* represents the year of data you want to obtain, *dataset* represents the dataset which you want to get [datasets](https://api.census.gov/data/2020/acs/acs5/variables.html), *variables* represent variables that you are interested in, and *location* represents the level of the data (i.e. the US, state, state counties). 

```python
# Request URL. 
query_url = f"{HOST}{year}{dataset}{g}{variables}{location}{KEY}"
```

The URL request for the Census Bureau is presented in the chuck of code above. It is worth noting that you will need to get an API key.

```python
# Send request to record the response from query_url.
response = requests.get(query_url)
```

```python
# Convert response to DataFrame.
df_income = json_to_dataframe(response)

df_income.head(1)
```

The table above represents the income per capita for the US states for the past 12 months, in 2020. Pennsylvania, for example, had an average per capita income of $35,518 in 2020, and 42 represent the FIPS code.

```python
# Group the dataset by party and states, 
#and sum up total votes party got. 
df_share = df_president[['party', "state", "total_votes"]]\
.groupby(['party', "state"], as_index = False).sum().sort_values(by='state', ascending = True)

# The subset of the dataset: accessing rows with REP or DEM party.
df_share = df_president[df_president\
                                  .party.isin(["REP","DEM"]) == True]

# Extracting variables for data cleaning: state, county, party, and total_votes.
df_share = df_share[["state","county",\
                                         "party","total_votes"]]


# Cleaning the data; switching rows to columns. 
df_share["REP_votes"] = df_share[df_share.\
                                 party == "REP"]["total_votes"]
df_share["DEM_votes"] = df_share[df_share.\
                                 party == "DEM"]["total_votes"]

# Grouping the data. 
#Mainly, we want to remove NAs in rows since REP total_votes does not have DEM total_votes.
df_share = df_share.groupby(["state"], as_index = False).sum()

# Creating new columns; percent of votes of each party from total votes.
df_share["REP_share"] = df_share["REP_votes"] / df_share["total_votes"] * 100
df_share["DEM_share"] = df_share["DEM_votes"] / df_share["total_votes"] * 100

# Rounding percent of votes.
df_share["REP_share"] = df_share["REP_share"]\
.round(decimals = 5)

df_share["DEM_share"] = df_share["DEM_share"]\
.round(decimals = 5)

# Changing the type of variables type from float to int
df_share = df_share.astype({"DEM_votes":'int', "REP_votes":'int'})
```

```python
# URL to the census.gov API.
# CONSTANT
HOST = "https://api.census.gov/data"

# Census API.
# CONSTANT
KEY = "&key={0}".format(census_api)

# URL struture for data extraction. 
# Year.
year = "/2021"

# Path to the required dataset.
dataset = "/pep/population"
g = "?get="

# Get variables required.
variables = "NAME,POP_2021"

# Geography code.
# "state:*" -> all states.
# "us:*" -> whole US.
# In general: "state:{state code}"
location = "&for=state:*"
```

```python
# Request URL. 
query_url = f"{HOST}{year}{dataset}{g}{variables}{location}{KEY}"
```

```python
# Send request to record the response from query_url.
response = requests.get(query_url)
```

<h5> Stepwise regression: Forward selection </h5>

```python
# Obtain regressors (independent variables).
X = regression.drop(["state", "county",
                     "state_number", "county_number",
                     "REP_share","DEM_share",
                     "population", "median_income"], axis=1).copy()
# Obtain response variable (dependent variable).
y = regression["REP_share"]
```

```python
def r2_adj(indep):
    """Return DataFrame with Adjusted R-squared for each regression,
    starting from one best explanatory variable and adding to it
    the next best regressor one at the time.
    """
    
    df_r2_adj = pd.DataFrame(columns=["Adjusted R-squared"])
    independent = indep[:] 
    r2_adj_selected = []
    
    while len(independent.columns) != 0:
        candidate = {}
        
        for i in independent:
            
            r2_adj_explanatory_variables = r2_adj_selected[:]
            r2_adj_explanatory_variables.append(i)
            r2_adj_reg = sm.OLS(y, sm.add_constant(regression[r2_adj_explanatory_variables])).fit()
            candidate[r2_adj_reg.rsquared_adj] = i  
            
        r2_adj_selected.append(candidate[max(candidate)])
        df_r2_adj.loc[candidate[max(candidate)]] = max(candidate)
        independent = independent.drop(columns=[candidate[max(candidate)]])
    
    return df_r2_adj

    
```

```python
r2_adj(X)
```

```python
def akaike(indep):
    """Return DataFrame with Akaike Information Criteria (AIC)
    for each regression, starting from
    one best explanatory variable and
    adding to it the next best regressor one at the time.
    """
    
    df_aic_selected = pd.DataFrame(columns=["AIC"])
    independent = indep[:] 
    aic_selected = []
           
    while len(independent.columns) != 0:
        candidate = {}
        
        for i in independent:
            
            aic_selected_explanatory_variables = aic_selected[:]
            aic_selected_explanatory_variables.append(i)
            aic_selected_reg = sm.OLS(y, sm.add_constant(regression[aic_selected_explanatory_variables])).fit()
            candidate[aic_selected_reg.aic] = i  
            
            
        aic_selected.append(candidate[min(candidate)])
        df_aic_selected.loc[candidate[min(candidate)]] = min(candidate)
        independent = independent.drop(columns=[candidate[min(candidate)]])
    
    return df_aic_selected
```

```python
akaike(X)
```

```python
def mse(indep):
    """Return DataFrame with Mean Square Error (MSE)
    for each regression, starting from
    one best explanatory variable and
    adding to it the next best regressor one at the time.
    """
            
    df_mse_selected = pd.DataFrame(columns=["MSE"])
    independent = indep[:] 
    mse_selected = []
    
    
    while len(independent.columns) != 0:
        candidate = {}
        
        for i in independent:
            
            mse_selected_explanatory_variables = mse_selected[:]
            mse_selected_explanatory_variables.append(i)
            mse_selected_reg = sm.OLS(y, sm.add_constant(regression[mse_selected_explanatory_variables])).fit()
            candidate[mse_selected_reg.mse_resid] = i  
            
        mse_selected.append(candidate[min(candidate)])
        df_mse_selected.loc[candidate[min(candidate)]] = min(candidate)
        independent = independent.drop(columns=[candidate[min(candidate)]])
    
    return df_mse_selected
```

<h4> Machine Learning: Regression Tree </h4>

---

**Model**: $$\text{REP_share} = \beta_0 + \beta_1 \text{log(income)}_i + \epsilon_i$$

**Objective Function**: $$\frac{1}{n} \Sigma^{n}_{i=1}(\text{REP_share}_i - \widehat{\text{REP_share}_i})^2 + \text{Regularization} = $$

$$ = \frac{1}{n} \Sigma^{n}_{i=1}(\text{REP_share}_i - \beta_0 - \beta_1 \text{log(income)}_i - \epsilon_i )^2 + \text{Regularization} $$

**Minimazation Problem**: $$ \min_{\beta_0,...,\beta_{\theta}} \frac{1}{n} \Sigma^{n}_{i=1}(\text{REP_share}_i - \beta_0 - \beta_1 \text{log(income)}_i - \epsilon_i )^2 + \text{Regularization}$$




The machine learning algorithm solves an objective minimization problem presented above. It identifies the most appropriate value for the beta parameter for each observation on the sample to minimize the difference between the actual observation and the predicted value. It accomplishes this by breaking down the full range of log(income) into smaller non-overlapping groups. If the group size is more than one, the approach is substantially the OLS method. We can assign distinct betas to each observation if our group size is one, resulting in a forecast error of zero.


The regularization parameter controls for two main characteristics: first, the maximum allowed error, the difference between an observable and predicted value, and second, the size of a tree, i.e., the maximum number of nodes and depth. The ccp_alpha (pruning parameter) controls the maximum allowable error, with a lower allowed error implying a higher minimum number of groups required. The greater the ccp_alpha value, the more nodes are pruned. Reducing the number of nodes causes the algorithm to divide the range of independent variables into larger groups. 

```python
# Obtain regressors (independent variables).
X = regression[["log_income"]].copy()
# Obtain response variable (dependent variable).
y = regression["REP_share"].copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=31, shuffle = True)
```

```python
mse_train = []
for i in range(1,21):
    model_1 = tree.DecisionTreeRegressor(max_depth=i, ccp_alpha = 0.000015).fit(X,y)
    y_train_pred= model_1.predict(X)
    mse_train.append(metrics.mean_squared_error(y, y_train_pred))



mse_test = []      
for i in range(1,21):
    model_2 = tree.DecisionTreeRegressor(max_depth=i, ccp_alpha = 0.000015).fit(X_train,y_train)
    y_test_pred = model_2.predict(X_test)
    mse_test.append(metrics.mean_squared_error(y_test, y_test_pred))
```

Secondly, we can compare different terminal node squared errors. 

```python
# First terminal node (left to right)
mse_squared_error_tree = 0.022

df_slr = regression[["REP_share","log_income"]].copy()
df_slr = df_slr.loc[df_slr["log_income"] <= 10.04]

df_slr["predicted"] = 2.084 - 0.142 * df_slr["log_income"]
df_slr["error"] = df_slr["REP_share"] - df_slr["predicted"]
df_slr["squared_error"] = df_slr["error"]**2


mse_squared_error_slr = sum(df_slr["squared_error"]) / len(df_slr)


df_mse = pd.DataFrame(columns=["MSE"])
df_mse.loc["MSE TREE"] = mse_squared_error_tree
df_mse.loc["MSE SLR"] = mse_squared_error_slr

df_mse
```

```python
# Sixth terminal node (left to right)
mse_squared_error_tree = 0.021

df_slr = regression[["REP_share","log_income"]].copy()
df_slr = df_slr.loc[(df_slr["log_income"] >= 10.508) &
                    (df_slr["log_income"] <= 10.929) &
                    (df_slr["log_income"] > 10.612)]

df_slr["predicted"] = 2.084 - 0.142 * df_slr["log_income"]
df_slr["error"] = df_slr["REP_share"] - df_slr["predicted"]
df_slr["squared_error"] = df_slr["error"]**2


mse_squared_error_slr = sum(df_slr["squared_error"]) / len(df_slr)


df_mse = pd.DataFrame(columns=["MSE"])
df_mse.loc["MSE TREE"] = mse_squared_error_tree
df_mse.loc["MSE SLR"] = mse_squared_error_slr

df_mse
```

We can see that the regression tree constantly outperforms simple linear regression. While Ordinary Least Squares tries to form a perfect linear fit for all the points minimizing Mean Square Error, the machine learning approach attempts to get a best-fit conditionally, minimizing Mean Square Error. Another cause for concern is that log per capita income and Republican vote share are relatively weakly correlated, suggesting that there may be no relationship.

Overall, we might state that if we had perfect data, i.e., experimental data, the regression tree would require a smaller sample size to produce robust and causal inference than OLS regression. On the other hand, with a large dataset, which is costly, OLS regression might provide a better approximation of the population parameter. Both methods have benefits and drawbacks; we expect OLS to outperform regression if we use the diff-in-diff method or the instrumental variable approach. We might also include more nodes or variables in the machine learning method, eliminating the Mean Square Error; however, most likely, the model will be overfitting, i.e., robust for training datasets but bad for validation sets or other datasets.
